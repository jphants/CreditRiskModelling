```python
import pandas as pd

train_df = pd.read_csv("../data/raw/cs-training.csv")
test_df = pd.read_csv("../data/raw/cs-test.csv")

train_df = train_df.drop(columns=["Unnamed: 0"])

X_train = train_df.drop(columns=["SeriousDlqin2yrs"])
y_train = train_df["SeriousDlqin2yrs"]
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

class IncomeMissingIndicator(BaseEstimator, TransformerMixin):

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        X["IncomeMissing"] = X["MonthlyIncome"].isna().astype(int)
        return X
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

preprocessor = Pipeline([
    ("income_missing", IncomeMissingIndicator()),
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

```python
import joblib

X_train_processed = preprocessor.fit_transform(X_train)
joblib.dump(preprocessor, "../models/preprocessor.joblib")
```
